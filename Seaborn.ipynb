#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### solve

Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots that can be created using Seaborn along with their typical use cases:

Scatter Plot (sns.scatterplot):

Use: Scatter plots are useful for visualizing the relationship between two continuous variables. They can reveal patterns, trends, and the concentration of data points.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

####
Histogram (sns.histplot):

Use: Histograms are used to visualize the distribution of a single variable. They provide insights into the underlying data distribution, including the frequency of values within specified bins.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example
tips = sns.load_dataset("tips")
sns.histplot(tips["total_bill"], bins=30, kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

Pair Plot (sns.pairplot):

Use: Pair plots are used for visualizing relationships between multiple variables in a dataset. It displays scatter plots for pairs of variables and histograms along the diagonal, making it easy to identify patterns and correlations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()
```

####
Violin Plot (sns.violinplot):

Use: Violin plots are similar to box plots but provide additional information by displaying the probability density of the data at different values. They are useful for visualizing the distribution of data across categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

####
These are just a few examples, and Seaborn offers a wide range of plots for various use cases. The library is designed to work seamlessly with pandas DataFrames and provides an aesthetically pleasing and informative visualization layer on top of Matplotlib.

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. Note: timepoint, signal, event, and region are columns in the fmri dataset.


#### solve
Certainly! To achieve this, you can use Seaborn's lineplot function with the "fmri" dataset. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset from Seaborn
fmri_data = sns.load_dataset("fmri")

# Create a line plot with different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set plot labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of Signal Over Time for Different Events and Regions")

# Show the plot
plt.show()
```

####
In this example:

x="timepoint" specifies the x-axis as the "timepoint" column.

y="signal" specifies the y-axis as the "signal" column.

hue="event" colors the lines based on the "event" column.

style="region" adds different line styles based on the "region" column.

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'. Note: pclass, age, and fare are columns in the titanic dataset.


#### solve

Certainly! You can use Seaborn's boxplot function to create box plots for the "titanic" dataset with different values of 'pclass'. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Create box plots for 'age' and 'fare' based on 'pclass'
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.tight_layout()
plt.show()
```

####
In this example:

x='pclass' specifies the x-axis as the "pclass" column.

y='age' and y='fare' specify the y-axes as the "age" and "fare" columns, respectively.

Two subplots are created side by side to display box plots for both 'age' and 'fare'.

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.


#### solve

The "diamonds" dataset is actually part of the Seaborn library, so you can use it directly without loading it separately. Here's an example of how to plot a histogram for the 'price' column in the "diamonds" dataset, with different colors for different 'cut' categories using the hue parameter:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with different colors for 'cut' categories
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack", bins=30)

# Set plot labels and title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")

# Show the plot
plt.show()
```

####
In this example:

x='price' specifies the 'price' column for the histogram.

hue='cut' colors the bars based on the 'cut' column.

multiple="stack" stacks the histograms for different 'cut' categories.

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


#### solve

Certainly! You can use the pairplot function from Seaborn to create a pair plot for the "iris" dataset, with different colors for different "species" using the hue parameter. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from Seaborn
iris_data = sns.load_dataset("iris")

# Create a pair plot with different colors for 'species'
sns.pairplot(iris_data, hue='species', markers=["o", "s", "D"])

# Set plot title
plt.suptitle("Pair Plot of Iris Dataset with Hue for Species", y=1.02)

# Show the plot
plt.show()
```

####
In this example:

hue='species' colors the points based on the 'species' column.

markers=["o", "s", "D"] sets different markers for each species to enhance distinguishability in the plot.


#### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

#### solve

Certainly! The "flights" dataset in Seaborn represents the number of passengers on flights between various cities and months. You can use a heatmap to visualize the data. Here's an example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from Seaborn
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='viridis', linewidths=.5)

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count on Flights (Heatmap)')

# Show the plot
plt.show()
```
